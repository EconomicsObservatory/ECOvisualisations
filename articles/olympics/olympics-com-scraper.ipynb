```python
def dataFrame2VizzuData(df):
    import ipyvizzu as vz
    dummy = []
    for c in df.columns:
        if c.lower()=='year':
            t = "dimension"
        elif (type(df[c].values[0]) == np.int64) | (type(df[c].values[0]) == np.float64):
            t = "measure"
        else:
            t = "dimension"
        if t=='measure':
            dummy.append(
                {
                    "name": c,
                    "values": [int(i) for i in df[c].fillna(1).values],
                    "type": t,
                }
            )
        else:
            dummy.append(
                {"name": c, "values": list(df[c].fillna("").astype(str).values), "type": t}
            )
    return vz.Data({"series": dummy})
```

```python
open('data/olympics-data-1896-2022-extended-all.json','w').write(json.dumps(dataFrame2VizzuData(dqn)))
open('data/olympics-data-1896-2022-extended-all.js','w').write('export var data = '+json.dumps(dataFrame2VizzuData(dqn)))
open('data/olympics-data-1896-2022-extended-countries.json','w').write(json.dumps(dataFrame2VizzuData(dn2)))
open('data/olympics-data-1896-2022-extended-countries.js','w').write('export var countries = '+json.dumps(dataFrame2VizzuData(dn2)))
open('data/olympics-data-1896-2022-extended-athletes.json','w').write(json.dumps(dataFrame2VizzuData(dqm)))
open('data/olympics-data-1896-2022-extended-athletes.js','w').write('export var athletes = '+json.dumps(dataFrame2VizzuData(dqm)))
open('data/olympics-data-1896-2022-extended-athletes-winter.json','w').write(json.dumps(dataFrame2VizzuData(dqm[dqm['gameSeason']=='Winter'])))
open('data/olympics-data-1896-2022-extended-athletes-winter.js','w').write('export var athletes = '+json.dumps(dataFrame2VizzuData(dqm[dqm['gameSeason']=='Winter'])))
open('data/olympics-data-1896-2022-extended-countries-winter.json','w').write(json.dumps(dataFrame2VizzuData(dn2[dn2['gameSeason']=='Winter'])))
open('data/olympics-data-1896-2022-extended-countries-winter.js','w').write('export var athletes = '+json.dumps(dataFrame2VizzuData(dn2[dn2['gameSeason']=='Winter'])))
```
